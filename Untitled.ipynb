```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# Check the backend being used
print("Matplotlib backend:", matplotlib.get_backend())

# Simulation parameters
years = 80  # Total lifecycle years
working_years = 40  # Working period
initial_wealth = 0  # Starting wealth
initial_income = 1  # Starting annual income
risk_free_rate = 0.02  # Risk-free rate
equity_return = 0.06  # Expected equity return
equity_volatility = 0.2  # Equity volatility
relative_risk_aversion = 5  # Risk aversion coefficient

# Functions for scenarios
def optimal_investment_share(wealth, age):
    """Optimal investment strategy."""
    return min(max((1 / relative_risk_aversion) * (equity_return - risk_free_rate) / equity_volatility ** 2, 0), 1)

def hundred_minus_age_rule(age):
    """100-minus-age rule for equity allocation."""
    return max((100 - age) / 100, 0)

# Containers for results
wealth_optimal = np.zeros(years)
wealth_100_minus_age = np.zeros(years)
wealth_undersaving = np.zeros(years)

# Initial conditions
wealth_optimal[0] = initial_wealth
wealth_100_minus_age[0] = initial_wealth
wealth_undersaving[0] = initial_wealth

print("Starting simulation...")

# Simulation for all scenarios
for year in range(1, years):
    age = year
    if year <= working_years:
        income = initial_income
        savings_rate_optimal = 0.2
        savings_rate_undersaving = 0.1  # Adjusted for undersaving
    else:
        income = 0
        savings_rate_optimal = savings_rate_undersaving = 0

    # Optimal strategy
    invest_optimal = optimal_investment_share(wealth_optimal[year - 1], age)
    equity_return_actual = np.random.normal(equity_return, equity_volatility)
    wealth_optimal[year] = wealth_optimal[year - 1] * (1 + risk_free_rate + invest_optimal * (equity_return_actual - risk_free_rate)) + income * (1 - savings_rate_optimal)

    # 100-minus-age rule
    invest_100_minus_age = hundred_minus_age_rule(age)
    wealth_100_minus_age[year] = wealth_100_minus_age[year - 1] * (1 + risk_free_rate + invest_100_minus_age * (equity_return_actual - risk_free_rate)) + income * (1 - savings_rate_optimal)

    # Undersaving scenario
    wealth_undersaving[year] = wealth_undersaving[year - 1] * (1 + risk_free_rate + invest_optimal * (equity_return_actual - risk_free_rate)) + income * (1 - savings_rate_undersaving)

print("Simulation completed.")

# Plot wealth trajectories
plt.figure(figsize=(10, 6))
plt.plot(wealth_optimal, label="Optimal Strategy Wealth")
plt.plot(wealth_100_minus_age, label="100 Minus Age Wealth")
plt.plot(wealth_undersaving, label="Undersaving Scenario Wealth")
plt.title("Lifecycle Wealth Trajectories")
plt.xlabel("Age")
plt.ylabel("Wealth")
plt.legend()
plt.grid()

# Display the plot
plt.show()
```
